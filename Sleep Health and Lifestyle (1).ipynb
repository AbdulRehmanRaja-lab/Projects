### TASK 2 How do Physical Activity Levels impact Sleep Quality?
##### - Compare average sleep quality for individuals with different activity levels

```python
df
```

```python
grouped = df.groupby('Quality of Sleep')['Physical Activity Level'].mean().reset_index()
grouped
```

```python
grouped.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped['Physical Activity Level'], grouped['Quality of Sleep'], color='skyblue')
plt.title('Average Sleep Quality by Physical Activity Level', fontsize=16)
plt.xlabel('Physical Activity Level', fontsize=14)
plt.ylabel('Average Quality of Sleep', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

##### Conclusion - Higher the physical activity ,higher the quality of sleep, hence according to this data physical activity does play a mojor role in the quality of sleep
